## DataFrame Plot

1. Użyj danych z sekcji "Input" (patrz poniżej)
1. Wczytaj dane z ``DATA`` jako ``sensors: pd.DataFrame``
1. Wybierz arkusz ``Luminance``
1. Sparsuj kolumny z datami
1. Wybierz pożądaną datę i lokację, następnie próbkuj co godzinę
1. Aktywność jest gdy ``Luminance`` jest różna od zera
1. Wyświetl wykres (line) z godzinami aktywności w dla lokacji "Sleeping Quarters upper"
1. Dla dnia 2019-09-28

Podpowiedź:
* ``pd.Series.apply(np.sign)``
* ``pd.Series.resample('H').sum()``

```python
DATA = 'https://python.astrotech.io/_static/sensors-optima.xlsx'
```

```python
sensors = pd.read_excel(
    io=DATA,
    parse_dates=['datetime'],
    sheet_name='Luminance',
    header=1,
    index_col=0)
```

```python
location = 'Sleeping Quarters upper'
date = '2019-09-28'

activity = (sensors
            .loc[sensors['location'] == location]
            .loc[date, 'value']
            .apply(np.sign)
            .resample('H')
            .sum())

ax = activity.plot(color='red', figsize=(16,5))
_ = ax.set_yticks([0, 1])
_ = ax.set_yticklabels(['sleep', 'awake'])
```

```python
import numpy as np
import pandas as pd


DATA = 'https://python.astrotech.io/_static/sensors-optima.xlsx'


df = pd.read_excel(
    io=DATA,
    parse_dates=['datetime'],
    sheet_name=['Luminance'],
    header=1,
    index_col=0,
)

df = df['Luminance']
df['value'] = df['value'].apply(np.sign)
where = df['location'] == 'Sleeping Quarters upper'
date = df['date'] == '2019-09-28'
activity = df.loc[where & date, 'value'].resample('H').sum()

ax = activity.plot(kind='line', color='red', label='')
ax.set_yticks([0, 1])
ax.set_yticklabels(['sleep', 'awake'])
```
