## Import SQL

- łączenie do bazy danych
- wybór danych za pomocą SQL
- opcje importu

```python
%pip install --upgrade requests
```

```python
DATABASE = r'/tmp/apollo-timeline.sqlite3'
DB_APOLLO = 'https://github.com/AstroMatt/book-python/raw/master/numerical-analysis/pandas/data/apollo-timeline.sqlite3'
```

```python
import requests


with open(DATABASE, mode='wb') as file:
    resp = requests.get(DB_APOLLO)
    file.write(resp.content)
```

```python
import sqlite3


with sqlite3.connect(DATABASE) as db:
    for row in db.execute('SELECT * FROM logs'):
        print(row)
```

```python
SQL = """
    SELECT
        id,
        datetime,
        level,
        message
    FROM logs
    WHERE datetime > '1969-07-21'
    ORDER BY datetime ASC
    LIMIT 5
"""
```

```python
with sqlite3.connect(DATABASE) as db:
    df = pd.read_sql(SQL, db)
```
