# NumPy - Import i eksport - lekcja

```python
import numpy as np

URL = 'https://python.astrotech.io/_static/iris.csv'
```

## Wczytywanie danych
- z pliku
- z sieci

- `np.loadtxt()`: `.csv`, `.txt`, `.dat`
- `np.load()`: `.npy`
- `np.loads()`: `.pkl`
- `np.fromstring()`: ...
- `np.fromregex()`: ...
- `np.genfromtxt()`: ...

```python
X = np.loadtxt(URL, skiprows=1, delimiter=',', usecols=(0,1,2,3))
```

```python
y = np.loadtxt(URL, skiprows=1, delimiter=',', usecols=4, dtype=np.str)
species = np.unique(y)

y[ y == 'setosa' ] = 0
y[ y == 'virginica' ] = 1
y[ y == 'versicolor' ] = 2
y = y.astype(np.int8)
```

```python
header = np.loadtxt(URL, max_rows=1, delimiter=',', dtype=np.str, usecols=(0,1,2,3))
```

## Eksport

- plik tekstowy `np.savetxt()`: `.csv`, `.txt`, `.dat`
- plik binarny: `np.save()`: `.npy`
- plik binarny: `np.savez()`: `.npz`
- plik binarny: `np.savez_compressed()`: `.npz`

```python
FILE = r'iris.csv'
np.savetxt(FILE, X, fmt='%.2f', delimiter=',')

%cat $FILE
```

## Łączenie Tablic

- `np.concatenate()`

```python
setosa = (y == 0)
sepal_length = (header == 'sepal_length')
sepal_width = (header == 'sepal_width')
```

```python
sl = X[setosa, sepal_length]
```

```python
sw = X[setosa, sepal_width]
```
